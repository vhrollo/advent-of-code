```python
path = 'input.txt'
data = [[el.replace('(', '').replace(')', '').replace('#', '') for el in lines.split()] for lines in open(path).read().splitlines()]
compass = {
    'U': (-1,  0),
    'D': (1,  0),
    'L': (0, -1),
    'R': (0,  1)
}
data = [[compass[el[0]], int(el[1]), el[2]] for el in data]
```

```python
matrix = [['#']]
pos = [0, 0]

def insert_horizontal(end):
    if end:
        matrix.append(['.' for _ in matrix[0]])
    else:
        matrix.insert(0, ['.' for _ in matrix[0]])
        pos[0] += 1

def insert_vertical(end):
    if end:
        for i in range(len(matrix)):
            matrix[i].append('.')
    else:
        for i in range(len(matrix)):
            matrix[i].insert(0, '.')
        pos[1] += 1

for line in data:
    dir, steps, hex_value = line
    for i in range(steps):
        pos = [pos[0] + dir[0], pos[1] + dir[1]]

        if pos[0] < 0:
            insert_horizontal(False)
        elif pos[0] >= len(matrix):
            insert_horizontal(True)
        elif pos[1] < 0:
            insert_vertical(False)
        elif pos[1] >= len(matrix[0]):
            insert_vertical(True)
        else:
            pass
        matrix[pos[0]][pos[1]] = '#'

for row in matrix:
    #print(' '.join(row))
    pass
```

```python
def isValid(screen, m, n, x, y):
    if x < 0 or x >= m \
            or y < 0 or y >= n \
            or screen[y][x] == '#':
        return False
    return True

def floodFill(screen, m, n, x, y):
    queue = []
    
    queue.append((y, x))
    
    while queue:
        currPixel = queue.pop()

        posY = currPixel[0]
        posX = currPixel[1]
         
        if isValid(screen, m, n,  posX + 1, posY):
            screen[posY][posX + 1] = '#'
            queue.append([posY, posX + 1])
         
        if isValid(screen, m, n, posX-1, posY):
            screen[posY][posX-1] = '#'
            queue.append([posY, posX-1])
         
        if isValid(screen, m, n, posX, posY + 1):
            screen[posY + 1][posX] = '#'
            queue.append([posY + 1, posX])
         
        if isValid(screen, m, n, posX, posY-1):
            screen[posY-1][posX] = '#'
            queue.append([posY-1, posX])
        

n = len(matrix)
m = len(matrix[0])
 
i = 1
x = matrix[i].index('#') + 1
y = i

floodFill(matrix, m, n, x, y)

count = 0
for i in range(0,n):
    for j in range(0,m):
        if matrix[i][j] == '#': count += 1
    #print(' '.join(matrix[i]))

print(count)
```

```python
compass = {
    '0': (0, 1),  # R
    '1': (1, 0),  # D
    '2': (0, -1),  # L
    '3': (-1, 0)   # U
}

other_data = [[compass[el[2][-1]],int(el[2][:-1], 16)] for el in data]
vertecies = []
edge = 0

pos = [0, 0]
for data in other_data:
    dir, steps = data
    dir = [dir[0] * steps, dir[1] * steps]
    pos = [pos[0] + dir[0], pos[1] + dir[1]]
    vertecies.append(pos.copy())
    edge += steps
```

```python
import numpy as np
def PolyArea(x,y):
    x = np.array(x, dtype=np.int64)
    y = np.array(y, dtype=np.int64)
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))

y = [el[0] for el in vertecies]
x = [el[1] for el in vertecies]

print(PolyArea(y,x) + edge/2 + 1)
```
