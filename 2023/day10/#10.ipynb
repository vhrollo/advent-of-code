```python
rows = data.splitlines()
matrix = [list(row) for row in rows] #matrix[rows][column]
print(len(matrix), len(matrix[0]))
s = None
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if (matrix[i][j] == 'S'):
            s = [i, j]

matrix[s[0]][s[1]] = '|'

north = [-1, 0]
south = [1, 0]
east = [0, 1]
west = [0, -1]

pipes = {
    '|': [north, south],
    '-': [east , west ],
    'L': [north, east ],
    'J': [north, west ],
    '7': [south, west ],
    'F': [south, east ],
    '.': [            ] 
}

pipe_matrix = [row[:] for row in matrix]
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        pipe_matrix[i][j] = '.'
```

```python
def search(pos, dir):
    return [pos[0] + dir[0], pos[1] + dir[1]]

def action(pos, last_pos):
    pipe = matrix[pos[0]][pos[1]]
    ways = list(pipes[pipe])
    if last_pos in ways:
        ways.remove(last_pos)
    new_pos = search(pos, ways[0])
    negated_ways = [-x for x in ways[0]]
    return new_pos, negated_ways
```

```python
pos = s
last_pos = south
counter = 0

while pos != s or counter == 0:
    pipe_matrix[pos[0]][pos[1]] = counter
    pos, last_pos = action(pos, last_pos)
    counter += 1

print(counter/2)
```

```python
doubled_pipe_matrix = [['o' for _ in range(2 * len(pipe_matrix[0]))] for _ in range(2 * len(pipe_matrix))]

for i in range(len(pipe_matrix)):
    for j in range(len(pipe_matrix[i])):
        doubled_pipe_matrix[i*2][j*2] = pipe_matrix[i][j]



pos = s
last_pos = south
counter = -1.0

while pos != s or counter == -1.0:
    doubled_pipe_matrix[pos[0]*2 + last_pos[0]][pos[1]*2 + last_pos[1]] = counter + 0.5
    pos, last_pos = action(pos, last_pos)
    counter += 1
```

```python
def isValid(screen, m, n, x, y):
    if x < 0 or x >= m \
            or y < 0 or y >= n \
            or screen[x][y] == None\
            or type(screen[x][y]) == int\
            or type(screen[x][y]) == float:
        return False
    return True

def floodFill(screen, m, n, x, y):
    queue = []
    
    queue.append([x, y])

    while queue:
        currPixel = queue.pop()

        posX = currPixel[0]
        posY = currPixel[1]
         
        if isValid(screen, m, n,  posX + 1, posY):
            screen[posX + 1][posY] = None
            queue.append([posX + 1, posY])
         
        if isValid(screen, m, n, posX-1, posY):
            screen[posX-1][posY]= None
            queue.append([posX-1, posY])
         
        if isValid(screen, m, n, posX, posY + 1):
            screen[posX][posY + 1]= None
            queue.append([posX, posY + 1])
         
        if isValid(screen, m, n, posX, posY-1):
            screen[posX][posY-1]= None
            queue.append([posX, posY-1])
        

m = len(doubled_pipe_matrix)
n = len(doubled_pipe_matrix[0])
 
x = 0
y = 0

floodFill(doubled_pipe_matrix, m, n, x, y)

count = 0
for i in range(0,m,2):
    for j in range(0,n,2):
        if doubled_pipe_matrix[i][j] == '.': count += 1
        #print(doubled_pipe_matrix[i][j], end =' ')
    #print()

print(count)
```
