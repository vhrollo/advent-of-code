```python
data = open('input.txt').read()
patterns = [pattern.splitlines() for pattern in data.split('\n\n')]
```

```python
def check(par, par2):
    return par == par2
    
def count(par, par2):
    count = 0
    for i in range(0, len(par)):
        if par[i] != par2[i]:
            count +=1
    return count == 1


def check_horisontal(pattern):
    reflection_length = []

    def get(index):
        return list(pattern[index])

    for index in range(1,len(pattern)):
        test1 = get(index)
        test2 = get(index-1)

        if count(test1,test2) or check(test1, test2):
            smudged = True
            left = index -1
            right = index
            
            while 0 <= left and right < len(pattern):
                l = get(left)
                r = get(right)

                if check(l,r):
                    left -= 1
                    right += 1
                elif count(l,r) and smudged:
                    smudged = False
                    left -= 1
                    right += 1
                else:
                    break
            
            if (left == -1 or right == len(pattern)) and not smudged:
                reflection_length.append((index-1, index))

    return reflection_length



def check_vertical(pattern):

    reflection_length = []

    def get(index):
        return [el[index] for el in pattern]

    for index in range(1,len(pattern[0])):
        test1 = get(index)
        test2 = get(index-1)

        if count(test1,test2) or check(test1, test2):
            smudged = True
            left = index -1
            right = index
            
            while 0 <= left and right < len(pattern[0]):
                l = get(left)
                r = get(right)

                if check(l,r):
                    left -= 1
                    right += 1
                elif count(l,r) and smudged:
                    smudged = False
                    left -= 1
                    right += 1
                else:
                    break
            
            if (left == -1 or right == len(pattern[0])) and not smudged:
                reflection_length.append((index-1, index))

    return reflection_length


sum = 0
for pattern in patterns:
    print('-------------------------')
    h = check_horisontal(pattern)
    v = check_vertical(pattern)
    print(h,v)
    if len(h) != 0:
        sum += h[0][1]*100
    if len(v) != 0:
        sum += v[0][1]



print(sum)
```
