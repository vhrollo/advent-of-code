```python
path = 'input.txt'
data = open(path).read().split('\n\n')
rules = {}
for el in data[0].splitlines():
    el = el.replace('}', '').split('{')
    value = [[i for i in v.split(':')] for v in el[1].split(',')]
    rules[el[0]] = value

ratings = [e[1:-1].split(',') for e in data[1].splitlines()]
```

```python
#part 1
sumg = 0 
for rating in ratings:
    step = 'in'
    while step != 'A' and step != 'R':
        for val in rating:
            exec(val)

        rule = rules[step]
        for i in rule:
            if len(i) == 1:
                step = i[0]
                break
            else:
                if eval(i[0]):
                    step = i[1]
                    break
    
    if step == 'A':
        sumg += x+m+a+s

sumg
```

```python
#part 2 - interval theory
def apply_operator(value, operator, operand):
    operators = {
        '<': lambda x, y: x < y,
        '>': lambda x, y: x > y
    }
    return operators[operator](value, operand)


def range_split(rating ,step = 'in', index = 0):
    if step == 'A':
        return [rating]
    if step == 'R':
        return []
    rule = rules[step][index]

    if len(rule) == 1:
        return range_split(rating, step = rule[0])
    
    val = rule[0][0]
    operator = rule[0][1]
    mid_point = int(rule[0][2:])


    start = rating[val][0]
    stop = rating[val][1]


    start_eval = apply_operator(start, operator, mid_point)
    stop_eval = apply_operator(stop, operator, mid_point)

    if start_eval == stop_eval:
        if start_eval:
            return range_split(rating, step = rule[1])
        
        else:
            return range_split(rating, step = step, index = index+1)
    

    second_rating = rating.copy()
    if operator == '<':
        nstop = mid_point - 1
        nstart = mid_point
        rating[val] = [start, nstop]
        second_rating[val] = [nstart, stop]
        return range_split(rating, step=rule[1]) + range_split(second_rating, step=step, index=index + 1)
    else:
        nstop = mid_point
        nstart = mid_point + 1
        rating[val] = [nstart, stop]
        second_rating[val] = [start, nstop]
        return range_split(rating, step=rule[1]) + range_split(second_rating, step=step, index=index + 1)



combinations = 0
for range in range_split( {'x': [1, 4000], 'm': [1, 4000], 'a': [1, 4000], 's': [1, 4000]} ):
    current = 0
    current += range['x'][1] - range['x'][0] + 1
    current *= range['m'][1] - range['m'][0] + 1
    current *= range['a'][1] - range['a'][0] + 1
    current *= range['s'][1] - range['s'][0] + 1
    combinations += current

combinations
```
