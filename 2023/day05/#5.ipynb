```python
for i, a in enumerate(arrs2):
    # Sort on source
    new_order = np.argsort( a[:,1] )
    arrs2[i] = a[new_order]
locations = []

for seed in seeds:
    next = seed
    for a in arrs2:
        # Search in our sorted sources
        index = a[:,1].searchsorted(next, side='right')

        if index == 0: #  I.e, before the first source entry
            pass
        elif next <= (a[index-1,1] + a[index-1,2]): # Within the previous elements range
            next = a[index-1,0] + next - a[index-1,1]
        else:
            pass
    locations.append(next)

print( "And the solution is:", min(locations))
```

```python
inf = float('inf')

def range_split(start, stop, lut): 
    starts = lut[:,1]
    stops  = lut[:,1]+lut[:,2]-1
    dst = lut[:,0]

    index = starts.searchsorted(start,side='right')
    i1 = index-1

    nstop = min(
    stop,
    stops[i1] if (index>0 and start<stops[i1]) else inf,
    starts[index] if (index<len(lut)) else inf
    )

    translated = [(start, nstop)]
    if index > 0 and start <= stops[i1]: # in_translated_interval:
        translated = [(dst[i1] + start - starts[i1], dst[i1] + nstop - starts[i1])]

    if nstop < stop:
        # "Stepping in deeper" now with the new start one above the current interval
        return translated + range_split(nstop+1, stop, lut)
    else:
        return translated








rs = seeds.reshape((-1,2))
# change arrays to intervals rather than ranges
rs[:,1] = rs[:,0] + rs[:,1] - 1




for ar in arrs2:
    nrs = list()
    for r in rs:
        nrs+=range_split( r[0],r[1], ar )
    rs = nrs

min(rs)[0]
```
